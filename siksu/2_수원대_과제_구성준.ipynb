## 과제 1
사망자와 생존자의 비율을 원 그래프로 나타내기

### 사망자와 생존자 수 확인

```python
df['Survived'].value_counts()
```

- 답안

```python
x = df['Survived'].value_counts().index
y = df['Survived'].value_counts().values

plt.figure(dpi = 150)
plt.title('사망 VS 생존')

plt.pie(y, labels = x)
plt.legend(loc = 'best')
plt.show()
```

- 비율 확인

```python
x = df['Survived'].value_counts().index
y = df['Survived'].value_counts().values

plt.figure(dpi = 150)
plt.title('사망 VS 생존 \n탑승객 생존 유무 (0: 사망, 1: 생존)')

plt.pie(y, labels = x, autopct = '%.1f%%')
plt.legend(loc = 'best')
plt.show()
```

## 과제 2

### 객실 등급 비율확인

```python
df['Pclass'].value_counts()
```

- 답안

```python
x = df['Pclass'].value_counts().index
y = df['Pclass'].value_counts().values

plt.figure(dpi = 150, figsize=(5,5))
plt.title('객실 등급 분포')

plt.pie(y, labels = x)
plt.legend(loc = 'best')
plt.show()
```

## 과제 3
객실 등급별 객실 수를 막대그래프로 나타내기

- 답안

```python
x = df['Pclass'].value_counts().index
y = df['Pclass'].value_counts().values

plt.figure(dpi = 150,figsize=(5,5))
plt.xticks(x)

plt.title('객실 등급 분포')
plt.xlabel('객실 등급')
plt.ylabel('객실 수')

plt.bar(x,y)
plt.show()
```

- 개수 표시

```python
x = df['Pclass'].value_counts().index
y = df['Pclass'].value_counts().values

plt.figure(dpi = 150,figsize=(5,5))
plt.xticks(x)

plt.title('객실 등급 분포')
plt.xlabel('객실 등급')
plt.ylabel('객실 수')

for i in range(len(x)):
    height = y[i]
    plt.text(x[i], height + 0.1, '%.0f' %height, ha='center', va='bottom', size = 10)
plt.bar(x,y)
plt.show()
```

## 과제 4
객실 등급별 티켓 가격을 막대 그래프로 나타내기

- 답안

```python
x = df.groupby('Pclass').mean().sort_values(by = 'Fare', ascending = False).index
y = df.groupby('Pclass').mean().sort_values(by = 'Fare', ascending = False)['Fare']

plt.figure(dpi = 150, figsize=(5,5))

plt.title('객실 등급 별 가격')
plt.xlabel('객실 등급')
plt.ylabel('가격')
plt.xticks(x)

plt.bar(x,y)
plt.show()
```

- 가격 표시

```python
x = df.groupby('Pclass').mean().sort_values(by = 'Fare', ascending = False).index
y = df.groupby('Pclass').mean().sort_values(by = 'Fare', ascending = False)['Fare']

plt.figure(dpi = 150, figsize=(5,5))

plt.title('객실 등급 별 가격')
plt.xlabel('객실 등급')
plt.ylabel('가격')
plt.xticks(x)

for i in range(len(x)):
    plt.text(x[i], y[i+1], '%.2f' %y[i+1], ha='center', va='bottom', size = 10)

plt.bar(x,y)
plt.show()
```
